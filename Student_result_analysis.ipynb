## EDA(Exploratory Data Analysis)

```python
# Gender distribution
plt.figure(figsize=(5,5))
ax=sns.countplot(x="Gender",data=df)
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()

#From above chart,it is analysed that the number of females in the data is more than the number of males.
```

```python
#Groupby on the basis of parentEdu and calculating the average of marks of students

group=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(group)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(group,annot=True)
plt.title("Relationship b/w parent's education and student's score")
plt.show()

# from this visualisation, it is concluded that the education of parents have a good impact on their scores
```

```python
# Boxplot to show the outliers in Mathscore
sns.boxplot(x="MathScore",data=df)
plt.show()
```

```python
from scipy import stats

# Calculate Z-scores
z_scores = stats.zscore(df['MathScore'])

# Define a Z-score threshold for identifying outliers
z_threshold = 3

# Identify outliers using Z-scores
outliers_z = (z_scores > z_threshold) | (z_scores < -z_threshold)

# Remove outliers and create a new DataFrame without outliers
data_no_outliers_z = df[~outliers_z]

# Check the shape of the new DataFrame without outliers
print("Original data shape:", df.shape)
print("Data shape without outliers (Z-test):", data_no_outliers_z.shape)
```

```python
# Boxplot without outliers
sns.boxplot(x="MathScore",data=data_no_outliers_z)
plt.show()
```

```python
# Boxplot to represent the outliers in reading score
sns.boxplot(x="ReadingScore",data=df)
plt.show()
```
